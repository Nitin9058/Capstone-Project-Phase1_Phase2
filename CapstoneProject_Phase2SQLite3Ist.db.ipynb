```python
# import liabrary
import sqlite3
```

```python
# this is connect command for giving file name 
dbcon=sqlite3.connect('imbdIst_data_Database.db')
```

```python
cur=dbcon.cursor()
```

```python
# creating table
cur.execute("create table imbdmovie_Ist_table(Slno INT PRIMARY KEY,moviename TEXT,directorname TEXT,duration INT,ratings REAL)")
```

```python
with open('Table_imbd_Ist.csv','r') as file:
    total_records=0
    for row in file:
        cur.execute("insert into imbdmovie_Ist_table values(?,?,?,?,?)",row.split(","))
        dbcon.commit()
        total_records+=1
print(total_records,'Records Inserted')
```

```python
results=cur.execute("select * from imbdmovie_Ist_table")

for i in results:
    print(i)
```

```python
#Start Querying From First Table
# Query No 1
result=cur.execute("select * from imbdmovie_Ist_table where directorname in ('Christopher Nolan','Matt Reeves')")

for j in result:
    print(j)
```

```python
# Query No 2
result=cur.execute("select * from imbdmovie_Ist_table where duration>=140 and duration<=190")

for j in result:
    print(j)
```

```python
# Query No 3
result=cur.execute("select * from imbdmovie_Ist_table where ratings>7 order by ratings asc")

for j in result:
    print(j)
```

```python
# Query No 4
result=cur.execute("select * from imbdmovie_Ist_table order by moviename desc")

for j in result:
    print(j)
```

```python
# Query No 5
result=cur.execute("select * from imbdmovie_Ist_table where moviename like 'P%' and ratings>7")

for j in result:
    print(j)
```

```python
# this is connect command for giving second file name 
dbcon=sqlite3.connect('imbdmovieIInd_data_Database.db')
```

```python
# creating second table
cur.execute("create table imbdmovietablesecond(moviename TEXT,stars TEXT,genre TEXT,certificate TEXT,votes INT,grosscollection REAL,popularity REAL)")
```

```python
with open('Table_imbd_IInd.csv',encoding='utf-8') as file:
    total_records=0
    for row in file:
        cur.execute("insert into imbdmovietablesecond values(?,?,?,?,?,?,?)",row.split(","))
        dbcon.commit()
        total_records+=1
print(total_records,'Records Inserted')
```

```python
results=cur.execute("select * from imbdmovietablesecond")

for row in results:
    print(row)
```

```python
# Query No 1
result =cur.execute("select * from imbdmovietablesecond where stars ='Arnold Schwarzenegger'")

for j in result:
    print(j)
```

# Subquery

```python
sql = "SELECT imbdmovie_Ist_table.slno, imbdmovie_Ist_table.moviename, imbdmovie_Ist_table.directorname, imbdmovie_Ist_table.duration, imbdmovie_Ist_table.ratings, imbdmovietablesecond.stars, imbdmovietablesecond.genre, imbdmovietablesecond.certificate, imbdmovietablesecond.votes, imbdmovietablesecond.grosscollection, imbdmovietablesecond.popularity FROM imbdmovie_Ist_table INNER JOIN imbdmovietablesecond ON imbdmovie_Ist_table.moviename = imbdmovietablesecond.moviename"
result =cur.execute(sql)
for i in result:
    print(i)
```
